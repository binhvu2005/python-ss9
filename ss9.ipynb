```python
# Bài 1
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = x**2

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Biểu đồ đầu tay của Bạn")
ax.set_xlabel("Trục x")
ax.set_ylabel("Trục y")
ax.grid(True, linestyle='--')

plt.show()
```

```python
# Bài 2
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 200)
y = x**3

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x, y, color='red', linewidth=3)
ax.set_title("Đồ thị hàm bậc ba y = x³")
ax.set_xlabel("Giá trị x")
ax.set_ylabel("Giá trị y")
ax.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
# Bài 3
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = 2*x + 3

fig, ax = plt.subplots()
ax.plot(x, y, label='y = 2x + 3', color='blue', linewidth=3)
ax.set_title("Đường thẳng y = 2x + 3")
ax.set_xlabel("Trục x")
ax.set_ylabel("Trục y")
ax.legend()
ax.grid(True, linestyle='--')
plt.show()
```

```python
# Bài 4
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 200)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots()
ax.plot(x, y1, label='sin(x)', color='blue')
ax.plot(x, y2, label='cos(x)', color='red', linestyle='--')
ax.set_title("So sánh sin(x) và cos(x)")
ax.set_xlabel("Góc (radians)")
ax.set_ylabel("Giá trị hàm số")
ax.legend()
ax.grid(True, linestyle='--')
plt.show()
```

```python
# Bài 5
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y1 = 2 * x + 1
y2 = -x + 10
fig, ax = plt.subplots()
ax.plot(x, y1, label='đường 1', color='blue')
ax.plot(x, y2, label='đường 2', color='orange')

ax.set_title("Giao điểm của hai đường thẳng")
ax.set_xlabel("Trục x")
ax.set_ylabel("Trục y")
ax.legend()
ax.grid(True, linestyle='--')
plt.show()
```

```python
# Bài 6
import numpy as np
import matplotlib.pyplot as plt

years = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
population = [92, 93, 94, 95, 96, 97, 98, 99, 100, 101]

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(years, population, color='blue', linestyle='-', marker='o', label='Dân số')
ax.set_title("Tăng trưởng dân số Việt Nam 2015–2024")
ax.set_xlabel("Năm")
ax.set_ylabel("Dân số (triệu người)")
ax.set_xticks(years)
ax.grid(True, linestyle='--', alpha=0.7)
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# Bài 7
import numpy as np
import matplotlib.pyplot as plt

products = ['Áo', 'Quần', 'Giày', 'Túi', 'Mũ', 'Tất']
sales = [150, 120, 200, 90, 65, 80]

fig, ax = plt.subplots(figsize=(10, 6))
positions = np.arange(len(products))
bars = ax.bar(positions, sales, color='green', edgecolor='black', linewidth=1.2)
ax.set_xticks(positions)
ax.set_xticklabels(products)
ax.set_title("Doanh số bán hàng tháng 10/2025")
ax.set_ylabel("Doanh số (triệu)")

for bar in bars:
    h = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, h - h * 0.05, str(h), ha='center', va='center', color='white', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# Bài 8
import numpy as np
import matplotlib.pyplot as plt

months = np.arange(1, 13)
temperature = [20, 22, 25, 28, 30, 32, 31, 30, 28, 26, 23, 21]
rainfall = [10, 15, 30, 120, 200, 250, 280, 220, 150, 80, 30, 15]

fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(months, temperature, color='red', linestyle='-', marker=None, label='Nhiệt độ (°C)')
ax1.set_xlabel('Tháng')
ax1.set_ylabel('Nhiệt độ (°C)', color='red')
ax1.tick_params(axis='y', labelcolor='red')
ax1.set_xticks(months)
ax1.grid(True, linestyle='--', alpha=0.6)

ax2 = ax1.twinx()
ax2.scatter(months, rainfall, color='blue', s=80, label='Lượng mưa (mm)', linestyle='--')
ax2.set_ylabel('Lượng mưa (mm)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

ax1.set_title('Nhiệt độ và lượng mưa Hà Nội 2025')
plt.tight_layout()
plt.show()
```
